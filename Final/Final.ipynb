```python
# Final One zone-5  

import matplotlib.pyplot as plt
from math import *
import numpy as np

try:
    img = plt.imread("fish.png")
except FileNotFoundError:
    img = plt.imread("fish.png")

m, n = img.shape[:2]

angle = 215

# Here is the catch
alpha = (pi * (angle-180)) / 180

mrt = ceil(abs(m * cos(alpha)) + abs(n * sin(alpha)))
nrt = ceil(abs(m * sin(alpha)) + abs(n * cos(alpha)))
print(f"{(mrt, nrt) = }")

x_offset = ceil(m * sin(alpha))
fs = lambda y: y / tan(alpha + pi / 2)    # starting line equation

rot = np.zeros((mrt, nrt), dtype=img.dtype)

delta_x = abs(cos(alpha))
delta_y = abs(sin(alpha))

for i in range (ceil(m * cos(alpha))):
    x_, y_ = fs(i), i

    for px in range(n):
        try:
            x, y = floor(x_), floor(y_)
            
            rot[-y, -x - x_offset] = img[i, px]
            rot[-y - 1, -x - x_offset] = img[i, px]
            x_ += delta_x
            y_ += delta_y
        except IndexError:
            print(f"[index err] mapping {(x + x_offset, y)} to {(px, i)}")
            pass

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img, cmap="gray")

plt.subplot(1, 2, 2)
plt.title("Rotated Image")
plt.imshow(rot, cmap="gray")
plt.show()
```

```python
# Zone-6 ( 225 <= alpha <=270)
# Final
import matplotlib.pyplot as plt
from math import *
import numpy as np

try:
    img = plt.imread("fish.png")
except FileNotFoundError:
    img = plt.imread("fish.png")

m, n = img.shape[:2]

angle = 270

alpha = (pi *  (angle-180) ) / 180

mrt = ceil(abs(m * cos(alpha)) + abs(n * sin(alpha)))
nrt = ceil(abs(m * sin(alpha)) + abs(n * cos(alpha)))

x_offset = ceil(m * sin(alpha))
fs = lambda y: y / tan(alpha)                      

rot = np.zeros((mrt, nrt), dtype=img.dtype)

delta_x = abs(sin(alpha))
delta_y = abs(cos(alpha))

poopie = 0
for i in range(ceil(n * sin(alpha))):
    x_, y_ = fs(i), i
    
    for px in range(m):
        try:
            x, y, pp = floor(x_), floor(y_), round(poopie)
            rot[-y, -x - x_offset] = img[px, round(poopie)] # y, x
            rot[-y - 1, -x - x_offset] = img[px + 1, round(poopie)]
            x_ -= delta_x
            y_ += delta_y
        except IndexError:
            print(f"[index err] mapping {(x + x_offset, y)} to {(i, px)}")
            pass
    poopie += 1 / sin(alpha)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img, cmap="gray")

plt.subplot(1, 2, 2)
plt.title(f"{(round(delta_x, 3), round(delta_y, 3)) = }")
plt.imshow(rot, cmap="gray")
plt.show()
# plt.savefig(f"outputs/dlrz2_{str(angle).rjust(3, '0')}.png")
```
